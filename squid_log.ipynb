```python
#load libraries.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display, HTML
from matplotlib.gridspec import GridSpec

```

```python
#load data

logs_csv = pd.read_csv("logs.csv")
```

```python
#time conversion
logs_csv['time'] = pd.to_datetime(logs_csv['time'],unit='s')
logs_csv = logs_csv.drop(columns=logs_csv.columns[0])
# logs_csv.head(20)

# logs_csv = logs_csv.drop('time', 1)
```

```python
#creating the unique list of tables
logs_csv.nunique().reset_index().to_csv("unique_elements.csv")
```

```python
unique = pd.read_csv("unique_elements.csv")
unique
unique.drop(columns=unique.columns[0])
```

```python
# logs_csv['time'].dtype
logs_csv['date'] = logs_csv['time'].dt.date
logs_csv['main_time'] = logs_csv['time'].dt.time
```

```python
dataframe1 = logs_csv.drop('time', 1)

dataframe1.head(10)
```

```python
# dataframe1 = pd.to_datetime(dataframe1['date'])
# dataframe1
dataframe1['date'] = pd.to_datetime(dataframe1['date'])

# dataframe2 = dataframe1
# dataframe1['main_time']
# dataframe2['main_time'] = dataframe1['main_time'].str.split(".",expand= True)

# dataframe2['main_time'] = dataframe2['main_time'].str.split(".",expand = True)[1]
```

```python
dataframe1['main_time'].astype(str)

dataframe1['time'] = dataframe1['main_time'].astype(str).str.split(".",expand = True)[0]


```

```python
dataframe1.drop(columns='main_time')
```

```python
dataframe1['time'] = pd.to_datetime(dataframe1['time'],format = '%H:%M:%S')
dataframe1['hour'] = dataframe1['time'].dt.hour
# dataframe1['hour'] = dataframe1['time'].hour
# pd.to_datetime(s, format='%d%b%Y %H:%M')
```

```python
dataframe2 = dataframe1.drop(columns='main_time')
```

```python
dataframe2
# dataframe2['just_time'] = dataframe2['time'].hour

dataframe2
```

```python
dataframe2 = dataframe2.drop(columns='time')
dataframe2

# dataframe2.groupby('client_address').count().reset_index()
```

```python
#check the maximum amount of transacted data by client_address type

# dataframe2.groupby('client_address').sum().reset_index().sort_values(by = 'bytes',ascending = False).head(9).to_csv("Transaction_top3.csv")

top9_data = dataframe2.groupby('client_address').sum().reset_index().sort_values(by = 'bytes',ascending = False).head(9)

sns.barplot(x='bytes',y='client_address',data=top9_data)
```

```python
#check the maximum duration of connection by client_address type
top9_duration = dataframe2.groupby('client_address').sum().reset_index().sort_values(by = 'duration',ascending = False).head(9)

# top9_duration.to_csv("Duration.csv")

sns.barplot(x='duration',y='client_address',data=top9_duration)
```

```python
# dataframe2.groupby('request_method').sum().reset_index()

# dataframe2['request_method'] = dataframe2['request_method']
# dataframe2.request_method.unique()

dataframe2['URL'] = dataframe2['URL'].astype(str)

top9_URL = dataframe2.groupby('URL').sum().reset_index().sort_values(by = 'bytes',ascending = False).head(9)


sns.barplot(x='bytes',y='URL',data=top9_URL)
```

```python
# dataframe2['hours'] = dataframe2['just_time'].dt.hours
top_time = dataframe2.groupby('hour').sum().reset_index().sort_values(by = 'bytes',ascending = False).head(9)
sns.barplot(x='hour',y='bytes',data=top_time)
```

```python
top_traffic = dataframe2.groupby('hour').count().reset_index().sort_values(by = 'client_address',ascending = False).head(9)

top_traffic
sns.barplot(x='hour',y='client_address',data=top_traffic)
```
